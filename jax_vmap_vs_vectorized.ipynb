```python
import jax
import jax.numpy as np
import numpy as npo


N = 100


def single(key):
    x = jax.random.normal(key, shape=(1,))
    for i in range(10):
        x = 0.9*np.sin(x)
    return x

def vectorized(key):
    x = jax.random.normal(key, shape=(N,))
    for i in range(10):
        x = 0.9*np.sin(x)
    return x

def vmapped(key):
    keys = jax.random.split(key, N)
    return jax.vmap(single)(keys)

js = jax.jit(single)
print(npo.array(js(jax.random.PRNGKey(0))))

def vjmapped(key):
    keys = jax.random.split(key, N)
    return jax.vmap(js)(keys)
```

```python
N = 100

def vectorized(key):
    a = np.array([0.])
    def inner(a):
        x = jax.random.normal(key, shape=(N,))
        for i in range(10):
            x = 0.9*np.sin(x + a * np.ones_like(x))
        return x.sum()
    for j in range(10):
        a = a - 0.01 * jax.grad(inner)(a)
    return a

def vmapped(key):
    keys = jax.random.split(key, N)
    a = np.array([0.])
    def inner_with_key(key, a):
        x = jax.random.normal(key, shape=(1,))
        for i in range(10):
            x = 0.9*np.sin(x + a * np.ones_like(x))
        return x.sum()
    
    def inner(a):
        return jax.vmap(inner_with_key, (0, None))(keys, a).sum()

    for j in range(10):
        a = a - 0.01 * jax.grad(inner)(a)
    return a
```

```python
import time

tstart = time.time()
jvec = jax.jit(vectorized)
for i in range(100):
    print(npo.array(jvec(jax.random.PRNGKey(i))))
    print("iter {} time {}".format(i, time.time() - tstart))
tfin = time.time()
print("total time ", tfin - tstart)
```
